1. Explain what inheritance is in object-oriented programming and why it is used.

Ans-
Inheritance allows a new class to reuse the properties, behavior and methods of existing class.

Class whose properties and methods are inherited are known as parent class while the class that inherits the properties from parent class called ad child class.

2. Discuss the concept of single inheritance and multiple inheritance, highlighting their differences and advantages.

Ans-
Single Inheritance : It allows a new class to reuse the properties, behavior and methods from only one super class.

It promotes simplicity and avoids ambiguity in the code.

Multiple Inheritance : It allows a new class to reuse the properties, behavior and methods from two or more super class.

It allows for code re usability and flexibility.

Differences :

Both single Inheritance and multiple inheritance lies in the number of superclasses and a subclass can have, single Inheritance allows only one direct superclass, leading to a linear class hierarchy.

Multiple Inheritance permits a subclass to inherit from multiple direct superclasses. It creates more complex and tree like hierarchy.

Single inheritance promotes simplicity and ease of understanding.

Multiple Inheritance provides greater code reusability and flexibility .

4. What is the significance of the "protected" access modifier in inheritance? How does it differ from "private" and "public" modifiers?

Ans-
When we declare a class member as protected it means that this member is not accessible from outside the class hierarchy. Any derived class that inherits from this class can access and use the protected member.

protected class members are like a secret shared among a class and its subclasses, allowing them to work together and create specialized behaviors without exposing everything to the outside world.

Public class members can be accessed from anywhere. It can be used and modified by any part of code.

Private class members are only accessible with in the class where they are defined. It cannot be accessed and modified from outside the class.

Super keyword in Inheritance allows derived class to access and use method, properties of its base class or parent class.

class parent:

        self.first_name = first_name
        self.last_name = last_name

        print(f"Good Evening!, I am {self.first_name} {self.last_name}")

        super().__init__(first_name, last_name)

class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number =account_number
        self.balance = balance
        
class SavingsAccount(BankAccount):
    def calculate_interest(self, interest_r):
        interest = self.balance * interest_r
        self.balance = self.balance + interest
        print(f"{self.balance}")
        
class CheckingAccount(BankAccount):
    def deduct_fees(self, fee_amount):
        self.balance = self.balance - fee_amount
